```python
# Set plot style
sns.set_style('whitegrid')

# Set color palette
colors = ['#FF6384', '#36A2EB']  # Customize the colors as per your preference
sns.set_palette(sns.color_palette(colors))

# Define the features for bivariate analysis
features = ['Age', 'DailyNetSpent', 'DailySiteSpent']

# Create subplots dynamically based on the number of features
num_plots = len(features)
fig, axes = plt.subplots(1, num_plots, figsize=(16, 6))

# Perform bivariate analysis using a loop
for i, feature in enumerate(features):
    ax = axes[i]
    ax.set_title(f'{feature} vs. Click on Ad')
    sns.boxplot(x='ClickOnAd', y=feature, data=data, linewidth=2.5, width=0.6, notch=True, saturation=0.8, ax=ax)
    ax.set_xlabel('Clicked on Ad')
    ax.set_ylabel(feature)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
